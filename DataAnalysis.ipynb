```python
#Sessional trend : This is seasonal data  useful for seeing which months have higher crime rates, but not for forecasting month-by-month trends.
crimes_month = df.groupby('MONTH').size()
crimes_month
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x=crimes_month.index, y=crimes_month.values)
plt.xlabel("Month")
plt.ylabel("Crimes")
plt.title("Crimes Per MOnth")
plt.show()
```

```python
#  Time-based trend (month + year): This is a time series  useful for predicting future months.
# You know when each crime happened.

df['YEAR_MONTH'] = df['Date'].dt.to_period('M')
crimes_monthly = df.groupby('YEAR_MONTH').size()
crimes_monthly
```

```python
import matplotlib.pyplot as plt

# Monthly total across all years
crimes_month = df.groupby('MONTH').size()

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
crimes_month.plot(kind='bar', color='skyblue')
plt.title("Total Crimes by Month (All Years Combined)")
plt.xlabel("Month")
plt.ylabel("Crime Count")
plt.xticks(rotation=0)

# True monthly trend over time
df['YEAR_MONTH'] = df['Date'].dt.to_period('M')
crimes_monthly = df.groupby('YEAR_MONTH').size()

plt.subplot(1, 2, 2)
crimes_monthly.plot(color='orange')
plt.title("Monthly Crime Trend Over Time")
plt.xlabel("Year-Month")
plt.ylabel("Crime Count")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Crimes per day of week
df['DAY_OF_WEEK'] = df['Date'].dt.day_name()
plt.figure(figsize=(10,5))
sns.countplot(x='DAY_OF_WEEK', data=df, order=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
plt.title('Crimes by Day of Week')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Crimes per year
plt.figure(figsize=(10,5))
sns.countplot(x='YEAR', data=df)
plt.title('Crimes per Year')
plt.show()

# Crimes per month
plt.figure(figsize=(10,5))
sns.countplot(x='MONTH', data=df)
plt.title('Crimes per Month')
plt.show()

# crimes per hour
plt.figure(figsize=(12,5))
sns.countplot(x='HOUR', data=df)
plt.title('Crimes by Hour of Day')
plt.show()
```

```python
import folium # folium is a powerful python library used for visualising geospatial data on interactive maps
from folium.plugins import HeatMap

# Base map centered around dataset mean coordinates
crime_map = folium.Map(location=[df['Latitude'].mean(), df['Longitude'].mean()],
                    zoom_start=12, tiles="cartodbpositron")

# Prepare data for heatmap
heat_data = df[['Latitude', 'Longitude']].dropna().values.tolist()

# Add heatmap layer
HeatMap(heat_data, radius=8, blur=6, max_zoom=13).add_to(crime_map)

# Save to HTML
crime_map.save("crime_hotspots.html")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))  # Bigger plot

top_crimes = df['TYPE'].value_counts().head(5).index

markers = ['o', 's', '^', 'D', 'x']  # Different markers for each crime type

for i, crime in enumerate(top_crimes):
    temp = df[df['TYPE'] == crime].groupby('YEAR').size()
    temp.plot(marker=markers[i], linestyle='-', label=crime)

plt.title("Top 5 Crime Types Over Years", fontsize=16)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Number of Crimes", fontsize=14)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(title="Crime Type")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
